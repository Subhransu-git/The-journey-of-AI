# ploting Bar Chart

```python
# pip install matplotlib
# A bar chart is used for categorical data (names, labels, groups).
# used for comperison
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

#style are using for background color
```

```python
#data 
classes =["python","html","css","ML","DS"]


class1_students= [30,10,20,25,10] #out of 100 student in each class
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
plt.show()
```

```python
#barh is using for horizontal
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students, width = 0.2)
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge")
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r")
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black")
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5)
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.2)
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--")
# If you want to make the graph plot more transparent, then you can make alpha
```

```python
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens",visible=False)
```

```python
style.use("ggplot")
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens")
```

```python

plt.figure(figsize=(16,9))
plt.bar(classes,class1_students, width = 0.2,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens")
```

```python

plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens")
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens")
plt.title("Bar Chart")
plt.xlabel("Classes")
plt.ylabel("No of Students")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,align="edge" ,color="r",edgecolor="black" ,linewidth=5 ,alpha=0.9,linestyle="--",
        label="class 1 studens")
plt.title("Bar Chart",fontsize=25)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.show()
```

```python
#if you want to add class 1 class 2 class 3 students
#graph overlap
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,color="b",label="class 1  students")
plt.bar(classes,class2_students,width=0.2,color="r",label="class 2  students")

plt.title("Bar Chart",fontsize=25)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index =np.arange(len(classes))
# np.arange(5)

width =0.4
plt.bar(classes_index,class1_students,width,color="b",label="class 1  students")
plt.bar(classes_index,class2_students,width,color="r",label="class 2  students")

plt.title("Bar Chart",fontsize=25)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index =np.arange(len(classes))
width =0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1  students")
plt.bar(classes_index+width,class2_students,width,color="r",label="class 2  students")

plt.xticks(classes_index +width,classes)
plt.title("Bar Chart",fontsize=25)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index =np.arange(len(classes))
width =0.2
plt.bar(classes_index,class1_students,width,color="b",label="class 1  students")
plt.bar(classes_index+width,class2_students,width,color="r",label="class 2  students")
plt.bar(classes_index+width+width,class3_students,width,color="y",label="class 2  students")

plt.xticks(classes_index +width,classes,rotation=20)
plt.title("Bar Chart",fontsize=25)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No of Students",fontsize=25)
plt.show()
```

```python
 #horizontal
plt.figure(figsize=(16,9))
classes_index =np.arange(len(classes))
width =0.2
plt.barh(classes_index,class1_students,width,color="b",label="class 1  students")
plt.barh(classes_index+width,class2_students,width,color="r",label="class 2  students")
plt.barh(classes_index+width+width,class3_students,width,color="y",label="class 3  students")

plt.yticks(classes_index +width,classes,rotation=20)
plt.title("Bar Chart",fontsize=25)
plt.ylabel("Classes",fontsize=25)
plt.xlabel("No of Students",fontsize=25)
plt.legend()
plt.show()
```

# use bar chart when you want to compare different categories.

```python
import matplotlib.pyplot as plt

students = ["Rahul", "Aman", "Priya", "Neha"]
marks = [85, 92, 78, 88]

plt.bar(students, marks, color='orange')

plt.title("Marks Comparison of Students")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
