# Ploting Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
datascience_students_age = np.random.randint(15,45,(100))
javascripts_students_age = np.random.randint(18,45,(100))
```

```python
print(datascience_students_age)
print(javascripts_students_age)
```

# hist method use for histrograms

```python
plt.hist(datascience_students_age)
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

# bins

```python
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins)
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

# rwidth

```python
#0 to 1
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8)
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

# histtype

```python
#0 to 1
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='step')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

```python
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='bar',orientation='horizontal')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

# color

```python
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='bar',orientation='horizontal' ,color='r')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.show()
```

```python
bins=[15,20,25,30,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='bar',
         orientation='horizontal' ,color='r',label='datascience student')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.legend()
plt.show()
```

```python
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='bar',
         orientation='vertical' ,color='r',label='datascience student')
plt.hist(javascripts_students_age,bins,rwidth=0.8 ,histtype='bar',
         orientation='vertical' ,color='b',label='javascripts student')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.legend(loc=3)
plt.show()
```

```python
plt.hist(datascience_students_age,bins,rwidth=0.8 ,histtype='bar',
         orientation='vertical' ,color='r',label='datascience student')
plt.hist(javascripts_students_age,bins,rwidth=0.8 ,histtype='bar',
         orientation='vertical' ,color='b',label='javascripts student')
plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.legend(loc=3)
plt.show()
```

```python
plt.hist([datascience_students_age,javascripts_students_age],bins,rwidth=0.8 ,histtype='bar',
         orientation='vertical' ,color=['r','b'],label=['datascience student','javascripts student'])

plt.title("DATASCIENCE STUDENTS AGE")
plt.xlabel('student age category')
plt.ylabel('No. of Students age')
plt.legend(loc=3)
plt.show()
```

# Use histogram when you want to study the distribution of numerical data,


```python
import matplotlib.pyplot as plt

ages = [18, 17, 20, 21, 19, 23, 24, 25, 20, 21, 19, 18, 22, 23, 24]

plt.hist(ages, bins=4, color='skyblue', edgecolor='black')

plt.title("Distribution of Student Ages")
plt.xlabel("Age Groups")
plt.ylabel("Number of Students")

plt.show()
```
